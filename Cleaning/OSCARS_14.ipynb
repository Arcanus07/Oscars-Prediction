# Inflating values

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

data = pd.read_csv('oscars_12.csv')

data.head()

data.columns.tolist()

data.isnull().sum().tolist()

data.info()

data.describe()

data['ID'] = data.index

data

cols = data.columns.tolist()
cols = cols[-1:] + cols[:-1]
data = data[cols]

data.drop('Title', axis = 1, inplace = True)

data
```

```python
inf = pd.read_csv('inflate.csv')
inf.head()

inf['Money'][inf['Year'] == 2020]

data['Budget'] *= 20.35

data

key_year = inf['Year'].tolist()
value_money = inf['Money'].tolist()

dic = {} 
for key in key_year: 
    for value in value_money: 
        dic[key] = value
        value_money.remove(value)
        break  

data.tail()

for i in data.index:
    for key in dic.keys():
        if data['Year'][i] == key:
            data['Budget'][i] = data['Budget'][i]/dic[key]

data

x = data['Budget'][1]

x

data['Budget'] = data['Budget'].astype('int64')

data
```

```python
data = data[data['Budget'] <= 400000000]
```

```python
data.to_csv('ordered_oscars.csv', index = False)
```
