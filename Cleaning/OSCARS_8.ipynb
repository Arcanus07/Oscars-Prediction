```python
#using omdb api to get values od month, runtime, country, metascore, imdbrating
for k in keys[8000:]:
    try:
        base_url = "http://www.omdbapi.com/?t=" + k + "&apikey=ef77cff7"  

        response = requests.get(base_url)

        dic = eval(response.text)
        #month.append(dic['Released'])
        Runtime.append(dic['Runtime'])
        country.append(dic['Country'])
        Metascore.append(dic['Metascore'])
        imdbRating.append(dic['imdbRating'])
    except Exception as e:
        print(e)
        wrong_key.append(k)
        pass
```
