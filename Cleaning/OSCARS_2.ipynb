```python
movies = movies[movies['genres'] != '[]'] #Removing movies with empty genre
movies.info()
```

```python
movies = movies.reset_index() #Reset index
movies.drop(['index'], axis = 1, inplace = True)
```

```python
#Fixing the values of genre and storing them in Genre
for i in movies.index:
    dic = eval(movies['genres'][i]) 
    sum = ''
    for j in range(len(dic)):
        sum += dic[j]['name'] +' '
    movies['Genre'][i] = sum
```
