```python
import pandas as pd
```

Gettign data

```python
data = pd.read_csv('oscars_10')
```

```python
data.head()
```

```python
#Dropping old runtime columns as we have new runtime values obtained from API
data.drop('runtime', axis = 1, inplace = True)
```

```python
data.info()
```

```python
#Movies that won the oscars and have a Null Metascore
data[data['Metascore'].isnull() & data['Won'] == 1]
```

```python
data['imdbRating'].loc[7869] = 8.0
```

```python
data['Metascore'] = data['Metascore'].fillna(0)
```

```python
#Getting the average metascore of movies that won oscars

sum = 0.0
count = 0
for i in data.index:
    if data['is_Nominated'][i] == 1 and data['Won'][i] != 1 and data['Metascore'][i] != 0:
        sum += data['Metascore'][i]
        count += 1
```

```python
sum/count
```

```python
#Replacing the null metascore of movies that won the oscars with the above average
for i in data.index:
    if data['is_Nominated'][i] == 1 and data['Won'][i] != 1 and data['Metascore'][i] == 0:
        data['Metascore'][i] = 79.16
```

```python
data.dropna(inplace = True)
```

```python
data.reset_index(inplace = True)
```

```python
data.drop('index', axis =1 , inplace = True)
```

```python
data.info()
```

```python
#Removing all the words in Runtime and getting only the numbers
data['Runtime'] = data['Runtime'].str.extract('(\d+)', expand=False) 
data
```

```python
data['Runtime'] = data['Runtime'].astype('int64')
```

```python
data.info()
```

```python
#Gettign all the words and removing the numbers in month column
data['month'] = data['month'].str.replace('\d+', '')
```

```python
data
```

```python
data['month'].unique()
```

```python
#Assigning a number to each month in the order of distance from December 
month = {
    ' Jan ' : 11,
    ' Feb ' : 10,
    ' Mar ' : 9,
    ' Apr ' : 8,
    ' May ' : 7,
    ' Jun ' : 6,
    ' Jul ' : 5,
    ' Aug ' : 4,
    ' Sep ' : 3,
    ' Oct ' : 2,
    ' Nov ' : 1,
    ' Dec ' : 0
}
```

```python
data['months_till_Dec'] = 0
```

```python
data['months_till_Dec'] = data['month'].map(month)
```

```python
data['months_till_Dec'] = data['months_till_Dec'].astype('int64')
```

```python
data.drop('month', axis = 1, inplace = True)
```

```python
data.columns
```

```python
data = data[['title', 'budget', 'Runtime', 'months_till_Dec', 'Year', 'country',
             'Metascore', 'imdbRating', 'Animation', 'Comedy','Family', 'Adventure', 
             'Fantasy', 'Drama', 'Romance', 'Action', 'Crime','Thriller', 'History', 
             'Science', 'Fiction', 'Mystery', 'Horror', 'War','Foreign', 'Western', 
             'Documentary', 'Music', 'Biography', 'Sci-Fi','Film-Noir', 'Musical', 
             'Sport', 'is_Nominated', 'Won']]
```

```python
data
```

```python
data.rename(columns={'title': 'Title', 'budget': 'Budget', 'months_till_Dec' : 'Months_till_Dec', 'country' : 'Country'}, inplace=True)
```

```python
data['Country'].unique()
```

```python
#Doing what we did for Genre in previous notebooks
#That is, replace comas with space and then create dummy variables for each country
for i in data.index:
    data['Country'][i] = data['Country'][i].replace(',', '')
```

```python
for i in data.index:
    country = data['Country'][i].split()
    for c in country:
        if c not in data.columns:
            data[c] = 0
        data[c][i] = 1
```
